```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/health_dataset_changed.csv")
```

```python
df.head(5)
```

```python
from scipy import stats
from statsmodels.stats.weightstats import zconfint
from scipy.stats import ttest_ind, chi2_contingency, f_oneway

print("95% Confidence Interval")

numerical_cols = [
    'age',
    'weight',
    'bmi',
    'blood_pressure',
    'cholesterol_level',
    'glucose_level'
]

for col in numerical_cols:
    ci = zconfint(df[col])
    print(f"{col}: {ci}")
```

```python
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha/2)

print("\nCritical Value:", critical_value)
print("Significance Level:", alpha)
```

```python
smokers = df[df['smoking_status'] == 'Smoker']['bmi']
non_smokers = df[df['smoking_status'] == 'Non-Smoker']['bmi']

t_stat, p_value = ttest_ind(smokers, non_smokers)

print("\nT-Test Result")
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Accept H0")
```

```python
contingency_table = pd.crosstab(df['smoking_status'], df['diabetes'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nChi-Square Test")
print("Chi-Square Value:", chi2)
print("P-Value:", p)

if p < 0.05:
    print("Reject H0")
else:
    print("Accept H0")
```

```python

group1 = df[df['age_group'] == '18-25']['bmi']
group2 = df[df['age_group'] == '26-35']['bmi']
group3 = df[df['age_group'] == '36-45']['bmi']
group4 = df[df['age_group'] == '46-60']['bmi']
group5 = df[df['age_group'] == '60+']['bmi']

anova_result = f_oneway(group1, group2, group3, group4, group5)

print("\nANOVA Test")
print("F-Statistic:", anova_result.statistic)
print("P-Value:", anova_result.pvalue)

if anova_result.pvalue < 0.05:
    print("Reject H0")
else:
    print("Accept H0")
```

```python
covariance = np.cov(df['age'], df['bmi'])[0][1]
correlation = np.corrcoef(df['age'], df['bmi'])[0][1]

print("\nCovariance:", covariance)
print("Correlation:", correlation)
```
